```python
import random
from typing import Tuple, Optional

class GameError(Exception):
    """Base class for game-related exceptions"""
    pass

class InvalidInputError(GameError):
    """Exception raised for invalid user input"""
    pass

class Player:
    def __init__(self):
        self._score: int = 0
        self._number: Optional[int] = None

    def generate_number(self, min_value: int = 1, max_value: int = 100) -> None:
        """Generate a random number within the specified range"""
        try:
            if not isinstance(min_value, int) or not isinstance(max_value, int):
                raise TypeError("Range values must be integers")
            if min_value >= max_value:
                raise ValueError("Minimum value must be less than maximum value")
            self._number = random.randint(min_value, max_value)
        except (TypeError, ValueError) as e:
            raise GameError(f"Error generating number: {str(e)}")

    def get_number(self) -> int:
        """Get the player's current number"""
        if self._number is None:
            raise GameError("Number has not been generated yet")
        return self._number

    def increment_score(self) -> None:
        """Increment the player's score"""
        self._score += 1

    def get_score(self) -> int:
        """Get the player's current score"""
        return self._score

class Game:
    def __init__(self, num_rounds: int = 5):
        if not isinstance(num_rounds, int) or num_rounds <= 0:
            raise ValueError("Number of rounds must be a positive integer")

        self.NUM_ROUNDS = num_rounds
        self.current_round = 0
        self.player = Player()
        self.computer = Player()

    def display_welcome(self) -> None:
        """Display welcome message and game instructions"""
        print("\n" + "=" * 40)
        print("Welcome To Hasnain's Coding World")
        print("=" * 40)
        print("This is a High Low Guessing Game")
        print("Rules:")
        print("1. You'll get a random number")
        print("2. Guess if your number is higher or lower than computer's")
        print("3. Score points for correct guesses!")
        print("=" * 40 + "\n")

    def get_user_advice(self) -> str:
        """Get and validate user input"""
        max_attempts = 3
        attempts = 0

        while attempts < max_attempts:
            try:
                user_advice = input("Do you think that your number is higher or lower than the computer's? (high/low): ").lower().strip()
                if user_advice not in ["high", "low"]:
                    attempts += 1
                    remaining = max_attempts - attempts
                    if remaining > 0:
                        print(f"Invalid input! Please enter 'high' or 'low'. {remaining} attempts remaining.")
                    continue
                return user_advice
            except KeyboardInterrupt:
                raise GameError("\nGame interrupted by user")
            except Exception as e:
                raise GameError(f"Unexpected error during input: {str(e)}")

        raise InvalidInputError("Maximum attempts reached. Game over!")

    def check_guess(self, user_advice: str) -> Tuple[bool, int]:
        """Check if the user's guess is correct"""
        try:
            player_num = self.player.get_number()
            computer_num = self.computer.get_number()

            is_player_num_higher = player_num > computer_num
            is_advice_correct = (user_advice == "high" and is_player_num_higher) or \
                              (user_advice == "low" and not is_player_num_higher)

            return is_advice_correct, computer_num
        except GameError as e:
            raise GameError(f"Error checking guess: {str(e)}")

    def play_round(self) -> None:
        """Play a single round of the game"""
        try:
            self.current_round += 1
            print(f"\nRound: {self.current_round}")
            print("=" * 40)

            # Generate numbers for both player and computer
            self.player.generate_number()
            self.computer.generate_number()

            print(f"Your Number Is: {self.player.get_number()}")

            # Get user's guess
            user_advice = self.get_user_advice()

            # Check if guess is correct
            is_correct, computer_num = self.check_guess(user_advice)

            if is_correct:
                self.player.increment_score()
                print("\nðŸŽ‰ You are right! The computer's number is", computer_num)
            else:
                print("\nâŒ You are wrong! The computer's number is", computer_num)

            print(f"Your score is now {self.player.get_score()}")
            print("=" * 40)

        except InvalidInputError as e:
            print(f"\nError: {str(e)}")
            raise
        except GameError as e:
            print(f"\nGame Error: {str(e)}")
            raise
        except Exception as e:
            print(f"\nUnexpected error: {str(e)}")
            raise

    def provide_final_feedback(self) -> None:
        """Provide feedback based on final score"""
        try:
            score = self.player.get_score()
            print("\nGame Over!")
            print(f"Final Score: {score}/{self.NUM_ROUNDS}")

            if score == self.NUM_ROUNDS:
                print("ðŸ† Wow! You played perfectly!")
            elif score > self.NUM_ROUNDS // 2:
                print("ðŸ‘ Good job, you played really well!")
            else:
                print("ðŸ‘Š Better luck next time!")

        except Exception as e:
            print(f"\nError displaying final feedback: {str(e)}")

    def play_game(self) -> None:
        """Main game loop with error handling"""
        try:
            self.display_welcome()
            while self.current_round < self.NUM_ROUNDS:
                self.play_round()
            self.provide_final_feedback()

        except KeyboardInterrupt:
            print("\n\nGame interrupted by user. Thanks for playing!")
        except InvalidInputError as e:
            print(f"\nGame ended due to invalid input: {str(e)}")
        except GameError as e:
            print(f"\nGame ended due to error: {str(e)}")
        except Exception as e:
            print(f"\nGame ended due to unexpected error: {str(e)}")
        finally:
            print("\nFinal score:", self.player.get_score())

def main():
    try:
        game = Game()
        game.play_game()
    except ValueError as e:
        print(f"Error starting game: {str(e)}")
    except Exception as e:
        print(f"Unexpected error: {str(e)}")

if __name__ == "__main__":
    main()
```

# Another Version With Name Age And Choice Of Rounds And Complete error Handling

```python
import random
from typing import Tuple, Optional
from dataclasses import dataclass

@dataclass
class PlayerDetails:
    """Class to store player's personal details"""
    name: str
    age: int

    @classmethod
    def get_player_details(cls) -> 'PlayerDetails':
        """Get and validate player details from user input"""
        max_attempts = 3

        # Get name
        while True:
            name = input("Please enter your name: ").strip()
            if name and name.replace(" ", "").isalpha():
                break
            print("Please enter a valid name (letters only)")

        # Get age with validation
        for _ in range(max_attempts):
            try:
                age = int(input("Please enter your age: "))
                if 1 <= age <= 120:
                    return cls(name=name, age=age)
                print("Please enter a valid age between 1 and 120")
            except ValueError:
                print("Please enter a valid number for age")

        raise InvalidInputError("Maximum attempts reached for age input")

class GameError(Exception):
    """Base class for game-related exceptions"""
    pass

class InvalidInputError(GameError):
    """Exception raised for invalid user input"""
    pass

class Player:
    def __init__(self, details: Optional[PlayerDetails] = None):
        self._score: int = 0
        self._number: Optional[int] = None
        self.details = details

    def generate_number(self, min_value: int = 1, max_value: int = 100) -> None:
        """Generate a random number within the specified range"""
        try:
            if not isinstance(min_value, int) or not isinstance(max_value, int):
                raise TypeError("Range values must be integers")
            if min_value >= max_value:
                raise ValueError("Minimum value must be less than maximum value")
            self._number = random.randint(min_value, max_value)
        except (TypeError, ValueError) as e:
            raise GameError(f"Error generating number: {str(e)}")

    def get_number(self) -> int:
        """Get the player's current number"""
        if self._number is None:
            raise GameError("Number has not been generated yet")
        return self._number

    def increment_score(self) -> None:
        """Increment the player's score"""
        self._score += 1

    def get_score(self) -> int:
        """Get the player's current score"""
        return self._score

class Game:
    def __init__(self):
        self.player_details = None
        self.NUM_ROUNDS = self.get_num_rounds()
        self.current_round = 0
        self.player = Player()
        self.computer = Player()

    @staticmethod
    def get_num_rounds() -> int:
        """Get and validate the number of rounds from user input"""
        max_attempts = 3

        for attempt in range(max_attempts):
            try:
                rounds = int(input("How many rounds would you like to play? "))
                if rounds > 0:
                    return rounds
                print("Please enter a positive number of rounds")
            except ValueError:
                remaining = max_attempts - attempt - 1
                print(f"Please enter a valid number. {remaining} attempts remaining")

        raise InvalidInputError("Maximum attempts reached for rounds input")

    def initialize_game(self) -> None:
        """Initialize the game with player details"""
        try:
            self.player_details = PlayerDetails.get_player_details()
            self.player = Player(self.player_details)
        except InvalidInputError as e:
            raise GameError(f"Error initializing game: {str(e)}")

    def display_welcome(self) -> None:
        """Display welcome message and game instructions"""
        print("\n" + "=" * 40)
        player_name = self.player_details.name if self.player_details else "Player"
        print(f"Welcome {player_name} To Hasnain's Coding World")
        print("=" * 40)
        print("This is a High Low Guessing Game")
        print("Rules:")
        print("1. You'll get a random number")
        print("2. Guess if your number is higher or lower than computer's")
        print("3. Score points for correct guesses!")
        print(f"4. You'll be playing {self.NUM_ROUNDS} rounds")
        print("=" * 40 + "\n")

    def get_user_advice(self) -> str:
        """Get and validate user input"""
        max_attempts = 3
        attempts = 0

        while attempts < max_attempts:
            try:
                player_name = self.player_details.name if self.player_details else "Player"
                user_advice = input(f"{player_name}, do you think that your number is higher or lower than the computer's? (high/low): ").lower().strip()
                if user_advice not in ["high", "low"]:
                    attempts += 1
                    remaining = max_attempts - attempts
                    if remaining > 0:
                        print(f"Invalid input! Please enter 'high' or 'low'. {remaining} attempts remaining.")
                    continue
                return user_advice
            except KeyboardInterrupt:
                raise GameError("\nGame interrupted by user")
            except Exception as e:
                raise GameError(f"Unexpected error during input: {str(e)}")

        raise InvalidInputError("Maximum attempts reached. Game over!")

    def check_guess(self, user_advice: str) -> Tuple[bool, int]:
        """Check if the user's guess is correct"""
        try:
            player_num = self.player.get_number()
            computer_num = self.computer.get_number()

            is_player_num_higher = player_num > computer_num
            is_advice_correct = (user_advice == "high" and is_player_num_higher) or \
                              (user_advice == "low" and not is_player_num_higher)

            return is_advice_correct, computer_num
        except GameError as e:
            raise GameError(f"Error checking guess: {str(e)}")

    def play_round(self) -> None:
        """Play a single round of the game"""
        try:
            self.current_round += 1
            player_name = self.player_details.name if self.player_details else "Player"
            print(f"\nRound {self.current_round} - {player_name}'s turn")
            print("=" * 40)

            # Generate numbers for both player and computer
            self.player.generate_number()
            self.computer.generate_number()

            print(f"Your Number Is: {self.player.get_number()}")

            # Get user's guess
            user_advice = self.get_user_advice()

            # Check if guess is correct
            is_correct, computer_num = self.check_guess(user_advice)

            if is_correct:
                self.player.increment_score()
                print(f"\nðŸŽ‰ Well done {player_name}! The computer's number is", computer_num)
            else:
                print(f"\nâŒ Sorry {player_name}, that's wrong! The computer's number is", computer_num)

            print(f"Your score is now {self.player.get_score()}")
            print("=" * 40)

        except InvalidInputError as e:
            print(f"\nError: {str(e)}")
            raise
        except GameError as e:
            print(f"\nGame Error: {str(e)}")
            raise
        except Exception as e:
            print(f"\nUnexpected error: {str(e)}")
            raise

    def provide_final_feedback(self) -> None:
        """Provide feedback based on final score"""
        try:
            score = self.player.get_score()
            player_name = self.player_details.name if self.player_details else "Player"
            print("\nGame Over!")
            print(f"Final Score for {player_name}: {score}/{self.NUM_ROUNDS}")

            if score == self.NUM_ROUNDS:
                print(f"ðŸ† Incredible, {player_name}! A perfect score!")
            elif score > self.NUM_ROUNDS // 2:
                print(f"ðŸ‘ Well played, {player_name}! You've got great intuition!")
            else:
                print(f"ðŸ‘Š Keep practicing, {player_name}! You'll get better!")

        except Exception as e:
            print(f"\nError displaying final feedback: {str(e)}")

    def play_game(self) -> None:
        """Main game loop with error handling"""
        try:
            self.initialize_game()
            self.display_welcome()
            while self.current_round < self.NUM_ROUNDS:
                self.play_round()
            self.provide_final_feedback()

        except KeyboardInterrupt:
            print("\n\nGame interrupted by user. Thanks for playing!")
        except InvalidInputError as e:
            print(f"\nGame ended due to invalid input: {str(e)}")
        except GameError as e:
            print(f"\nGame ended due to error: {str(e)}")
        except Exception as e:
            print(f"\nGame ended due to unexpected error: {str(e)}")
        finally:
            if self.player_details:
                print(f"\nFinal score for {self.player_details.name}:", self.player.get_score())
            else:
                print("\nFinal score:", self.player.get_score())

def main():
    try:
        game = Game()
        game.play_game()
    except ValueError as e:
        print(f"Error starting game: {str(e)}")
    except Exception as e:
        print(f"Unexpected error: {str(e)}")

if __name__ == "__main__":
    main()
```
